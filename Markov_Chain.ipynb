```python
import numpy as np
import matplotlib.pyplot as plt

# Define states
states = ['A', 'B', 'C']

# Define transition probability matrix
# Rows: Current state, Columns: Next state
transition_matrix = np.array([
    [0.2, 0.6, 0.2],  # A -> [A, B, C]
    [0.1, 0.7, 0.2],  # B -> [A, B, C]
    [0.3, 0.3, 0.4]   # C -> [A, B, C]
])

# Initial state distribution
initial_state_dist = [0.5, 0.3, 0.2]

# Function to simulate Markov chain time series
def markov_chain_simulation(transition_matrix, initial_state_dist, num_steps):
    # Choose the initial state based on initial state distribution
    current_state = np.random.choice(states, p=initial_state_dist)
    
    # Store the states in a list
    state_sequence = [current_state]
    
    for _ in range(num_steps - 1):
        # Find the index of the current state
        current_state_index = states.index(current_state)
        
        # Choose the next state based on the transition matrix
        next_state = np.random.choice(states, p=transition_matrix[current_state_index])
        
        # Append the next state to the sequence
        state_sequence.append(next_state)
        
        # Move to the next state
        current_state = next_state
    
    return state_sequence

# Simulate the Markov chain for 100 steps
num_steps = 100
state_sequence = markov_chain_simulation(transition_matrix, initial_state_dist, num_steps)

# Display the state sequence as a time series
plt.figure(figsize=(10, 4))
plt.plot(state_sequence, marker='o', linestyle='-', color='b')
plt.title('Markov Chain Time Series')
plt.xlabel('Time Step')
plt.ylabel('State')
plt.grid(True)
plt.show()

# Display the first few states
print("First 20 steps in the Markov chain:")
print(state_sequence[:20])
```
